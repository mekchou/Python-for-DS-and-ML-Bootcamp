```python
import seaborn as sns
%matplotlib inline
```

```python
# import scipy to be able to add in pearsonr, which is removed from seaborn
import scipy.stats as stats
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# kde is the kde line
sns.displot(tips['total_bill'],kde=True, bins = 30)
```

```python
# workaround provided by https://stackoverflow.com/questions/52118245/python-seaborn-jointplot-does-not-show-the-correlation-coefficient-and-p-value-o

graph = sns.jointplot(x= 'total_bill', y= 'tip', data = tips,kind= 'scatter')
r,p = stats.pearsonr(tips['total_bill'],tips['tip'])
phantom, = graph.ax_joint.plot(tips['total_bill'],tips['tip'], ls='', alpha = 0)
graph.ax_joint.legend([phantom],['r={:f}, p={:f}'.format(r,p)])
```

```python
sns.pairplot(tips,hue = 'sex', palette='cool', diag_kind = 'hist')
```
