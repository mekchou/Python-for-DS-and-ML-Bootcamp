```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# functional
plt.plot(x,y)
# plt.show()
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,1.8])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig2 = plt.figure()
axes1=fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('large plot')
axes2.plot(y,x)
axes2.set_title('small plot')
```

```python
fig,axes = plt.subplots(nrows =3, ncols=3)
# axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows =3, ncols=3)
# axes.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

## Figure Size and DPI

```python
# fig = plt.figure(nrows = 2, ncols = 1, figsize=(8,2)) #dpi can set dpi
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8,2)) #dpi can set dpi

# ax = fig.add_axes([0,0,1,1])
axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```
