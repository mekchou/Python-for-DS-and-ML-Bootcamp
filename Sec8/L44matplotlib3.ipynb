```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y,color='orange')
# linewidth can be replaced by lw, linestyle can be replaced by ls
ax.plot(x,y,color='#FF8C00',linewidth=3, alpha=0.5,linestyle='--',marker='o',markersize=20,markerfacecolor='blue',markeredgewidth = 3, markeredgecolor = 'green') #RGB Hex Code
```

## Control over axis appearance and plot range

```python
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
# ax.plot(x,y,color='orange')
# linewidth can be replaced by lw, linestyle can be replaced by ls
ax2.plot(x,y,color='#FF8C00',linewidth=2, alpha=0.5,linestyle='--')

ax2.set_xlim([0,1])
ax2.set_ylim([0,2])
```
